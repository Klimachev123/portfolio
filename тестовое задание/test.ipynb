```python
size = pd.pivot_table(data,
               index=["Размер"],
               values=["Текущий остаток, шт."],
               aggfunc=np.sum).sort_values(by='Текущий остаток, шт.', ascending=False)
size
```

```python
size = size.reset_index()
size
```

```python
plt.figure(figsize=(22, 7))
plt.title("Распределение размеров. Общий график")
sns.barplot(data=size, x="Размер", y="Текущий остаток, шт.")
plt.show()
```

```python
size_high = size.loc[size['Текущий остаток, шт.'] > 500]

plt.figure(figsize=(22, 7))
plt.title("Распределение размеров (более 500шт.)")
sns.barplot(data=size_high, x="Размер", y="Текущий остаток, шт.")
plt.show()
```

```python
size_low = size.loc[size['Текущий остаток, шт.'] <= 500]
plt.figure(figsize=(22, 7))
plt.title("Распределение размеров (менее 500шт.)")
sns.barplot(data=size_low, x="Размер", y="Текущий остаток, шт.")
plt.show()
```

```python
data['Закупка'] = data['Розничная цена']*0.2
data
```

```python
def revenue(row):
    if row['Выкупили, руб.'] == 0:
        return 0
    elif row['Выкупили, руб.'] != 0:
        return row['Выкупили, руб.'] - row['Закупка']
    
data['Прибыль'] = data.apply(revenue, axis=1)
pd.pivot_table(data,
               index=["Артикул WB"],
               values=["Прибыль"],
               aggfunc=np.sum).sort_values(by='Прибыль', ascending=False).head(10)  
```
